Here, we are trying to get the filenames from a folder in a FTP server such as MassIVE Repository. As an example, I have used the MSV000082312 ID here. The FTP download link for that ID in the repository is given as: ftp://massive.ucsd.edu/v01/MSV000082312/

```python
from ftplib import FTP
```

Here, provide the ID with the corresponding version

```python
ftp = FTP('massive.ucsd.edu')
ftp.login()  # Log in as anonymous
ftp.cwd('/v01/MSV000082312/')  # Change to the target directory
```

```python
files = ftp.nlst()  # List files in the directory
for file in files:
    print(file)

ftp.quit()
```

## Defining a function to retrieve filenames from a folder in a server

```python
def list_ftp_directory(server, directory):
    try:
        ftp = FTP(server)
        ftp.login()  # Log in as anonymous
        ftp.cwd(directory)  # Change to the target directory
        files = ftp.nlst()  # List files in the directory
        for file in files:
            print(file)
        ftp.quit()
    except Exception as e:
        print(f"Failed to list directory: {e}")
```

```python
server = 'massive.ucsd.edu'
```

```python
# Replace '/v01/MSV000082312/peak/' with the correct path if needed
list_ftp_directory(server, '/v01/MSV000082312/')
```

```python
# Going into 'peak folder'
list_ftp_directory(server, '/v01/MSV000082312/peak/')
```
